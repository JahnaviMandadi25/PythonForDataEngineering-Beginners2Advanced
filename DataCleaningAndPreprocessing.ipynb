1. Handling Null Values


```python
import pandas as pd  # Import pandas for data manipulation

df = pd.read_csv('/content/color_srgb.csv')  # Read CSV file into a DataFrame
df.head()  # Display the first five rows of the original DataFrame

new_df = df.dropna()  # Remove rows with missing values and store in a new DataFrame
new_df.head()  # Display the first five rows of the cleaned DataFrame

df.fillna(0, inplace=True)  # Replace all missing values in the original DataFrame with 0
df.head()  # Display the first five rows after filling missing values

df['Name'].fillna('N/A', inplace=True)  # Replace missing values in the 'Name' column with 'N/A'
df.head()  # Display the first five rows after updating the 'Name' column

df_new = df.drop_duplicates()  # Remove duplicate rows and store in a new DataFrame
df_new.head()  # Display the first five rows of the DataFrame after removing duplicates
```

Bar graph


```python
# Import necessary libraries
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For data visualization

# Load the data from a CSV file into a DataFrame
data = pd.read_csv('/content/sample_data/california_housing_test.csv')
df = pd.DataFrame(data)  # Create a DataFrame from the loaded data

# Plot the data using a bar chart
df.plot(x='longitude', y='latitude', kind='bar', legend=False)
# x='longitude': Specifies the column to use for the x-axis
# y='latitude': Specifies the column to use for the y-axis
# kind='bar': Specifies the type of plot (bar chart in this case)
# legend=False: Hides the legend to keep the plot clean

# Label the x-axis
plt.xlabel('longitude')  # Set the label for the x-axis
# Label the y-axis
plt.ylabel('latitude')  # Set the label for the y-axis
# Set the title of the plot
plt.title('long vs lati')  # Add a title to the plot

# Display the plot
plt.show()  # Render the plot on the screen
```

Line graph


```python
import pandas as pd
import matplotlib.pyplot as plt

# Correctly structured data
data = {'Product': ['A', 'B', 'C'],
        'Sales': [2345, 3334, 25541]}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the data
df.plot(x='Product', y='Sales', kind='line')

# Labeling the axes and title
plt.xlabel('Product')
plt.ylabel('Sales')
plt.title('Product Sales')

# Display the plot
plt.show()
```

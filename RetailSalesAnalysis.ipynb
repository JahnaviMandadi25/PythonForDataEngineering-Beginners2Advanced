```python
import pandas as pd  # Importing the pandas library for data manipulation
import matplotlib.pyplot as plt  # Importing matplotlib for data visualization
import numpy as np  # Importing numpy for numerical operations

class RetailSalesAnalyzer:  # Defining the RetailSalesAnalyzer class

  def __init__(self):  # Constructor to initialize the class
    self.data = pd.read_csv('/content/retail_sales.csv')  # Reading the retail sales data from a CSV file
    self.data['Date'] = pd.to_datetime(self.data['Date'])  # Converting the 'Date' column to datetime format

  def data_clean(self):  # Method to clean the data
    self.data.dropna(inplace=True)  # Dropping rows with missing values

  def total_sales_by_product(self):  # Method to calculate total sales by product
    return self.data.groupby('Product')['Sales'].sum()  # Grouping by 'Product' and summing the 'Sales' values

  def best_selling_product(self):  # Method to find the best-selling product
    return self.total_sales_by_product().sort_values(ascending=False).index[0]  # Sorting sales in descending order and returning the top product

  def average_daily_sales(self):  # Method to calculate average daily sales
    return self.data.groupby('Date')['Sales'].mean()  # Grouping by 'Date' and calculating the mean sales

  def plot_sales_trend(self):  # Method to plot the sales trend over time
    self.data.groupby('Date')['Sales'].sum().plot(kind='line')  # Grouping by 'Date' and plotting total sales over time
    plt.xlabel('Date')  # Label for the x-axis
    plt.ylabel('Total Sales')  # Label for the y-axis
    plt.title('Sales Trend')  # Title for the plot
    plt.show()  # Displaying the plot

  def plot_sales_by_product(self):  # Method to plot sales by product
    self.total_sales_by_product().plot(kind='bar')  # Plotting total sales for each product as a bar chart
    plt.xlabel('Product')  # Label for the x-axis
    plt.ylabel('Total Sales')  # Label for the y-axis
    plt.title('Sales by Product')  # Title for the plot
    plt.show()  # Displaying the plot
```

```python
analyzer = RetailSalesAnalyzer()  # Creating an instance of the RetailSalesAnalyzer class

print(analyzer.total_sales_by_product())  # Printing the total sales by product
print(analyzer.best_selling_product())  # Printing the best-selling product
print(analyzer.average_daily_sales())  # Printing the average daily sales

analyzer.plot_sales_trend()  # Plotting the sales trend over time
analyzer.plot_sales_by_product()  # Plotting the sales by product
```
